I. Kiểu dữ liệu trong Python

> Lệnh **type**(<**data**>) sẽ trả lại kiểu dữ liệu hiện thời của data




```python
type_1 = type(2.4)
type_2 = type('HaNoi')
type_3 = type(3 > 2)
type_4 = type(123)

#Test
print('type(2.4):     ', type_1)
print('type(\'HaNoi\'): ', type_2)
print('type(3 > 2):   ', type_3)
print('type(123):     ', type_4)
```

II. Biểu thức số học trong Python


> Một số quy tắc tính toán trong Python



n2.png


> Python có thể xử lý tính toán với số lớn



```python
#Example

print('Cộng\t\t1 + 2 = ', 1 + 2, '\n')

print('Trừ\t\t2 - 3 = ', 2 - 3, '\n')

print('Nhân\t\t3 * 4 = ', 3 * 4, '\n')

print('Chia nguyên\t4//3 = ', 4//3, '\n')  # Chia nguyên

print('Chia thực\t4/3 = ', 4/3, '\n')    # Chia thực

print('Lũy thừa\t2**10 = ', 2**10, '\n')

print('Mod\t\t5%3 = ', 5%3, '\n')
```
