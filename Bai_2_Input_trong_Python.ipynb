**I. Hàm nhập dữ liệu**


*1.  Hàm input()*


>  Để khởi tạo giá trị ban đầu cho biến, ta có thể khởi tạo bằng cách gán giá trị hoặc dùng lệnh nhập: **<tên biến> = input(<"Tiêu đề mong muốn">)**


> Tên biến được đặt theo quy tắc **identifier**. Tên biến chỉ cho phép chứa ký tự chữ, số, dấu gạch dưới.Tên biến phải bắt đầu là chữ cái và nên dễ nhớ.


> Tên biến không được trùng với từ khóa

    

*2.   Lưu ý khi dùng hàm nhập input()*
> Python tự convert KDL khi nhập. Tuy nhiên, nếu muốn giới hạn kdl khi dùng hàm input, có thể thêm kdl trước input và xử lý cùng exception



> **<tên biến> = <KDL>(input(<"Tiêu đề mong muốn">))**








```python
#Example
x = input('NHập x = ')
y = float(input('Nhập số thực y = '))
z = int(input('Nhập số nguyên z = '))
print('---------------------------')
print(type(x), x)
print(type(y), y)
print(type(z), z)
```

**II. Gán giá trị **


*1.   Gán giá trị*
> Để khởi tạo giá trị ban đầu cho biến, ngoài cách dùng hàm nhập, có thể dùng cách gán giá trị (hash code)
> Cú pháp 

>  **tên biến = value**

>Gán tuần tự 

var1 = value1

var2 = value2

.......................

>Hoặc gán liên tục 

var1, var2,... = value1, value, .....







```python
#Example1 - Gán biển 
x1 = 1
x2 = 2
x3, x4 = 1.222, "Nuoi_be_mo"

print(type(x1), x1)
print(type(x2), x2)
print(type(x3), x3)
print(type(x4), x4)

```

*2.Hàm swap*
>Hàm swap 

>Python hỗ trợ swap trực tiếp mà không cần viết hàm

var1, var2 = value1, value2

var1, var2 = var2, var1
>Ham swap có tác dụng đổi giá trị các biến

```python
#Example
x1, x2 = 'Python', 'C++'
print('x1_first: ', x1,'   ,x2_first: ', x2)
x1, x2 = x2, x1
print('x1_last:  ', x1,'      ,x2_last : ', x2)
```

```python
#Example
x1, x2, x3 = 'c', 'c++', 'Python'
print(' x1_first:',x1, '\n', 'x2_first:', x2, '\n','x3_first:', x3, '\n')
print('------------------------------------------------------------------')
x1, x2, x3 = x3, x1, x2
print(' x1_last: ', x1, '\n', 'x2_last: ', x2, '\n','x3_last: ', x3, '\n')
```
