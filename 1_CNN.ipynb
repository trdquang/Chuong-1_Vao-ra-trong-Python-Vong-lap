## Example

```python
import numpy as np

data = np.random.randint(0,5,(5,5))
print(data)
```

```python
import tensorflow as tf

data_tf = tf.convert_to_tensor(data, dtype=tf.float32)
print(data_tf)
```

```python
data_tf = tf.reshape(data_tf, (1, 5, 5, 1))
print(data_tf.shape)
```

```python
import tensorflow.keras as keras

conv_layer = keras.layers.Conv2D(filters=1, kernel_size=(3, 3))

# init
output = conv_layer(data_tf)
```

```python
weights = conv_layer.get_weights()
```

```python
print(weights[0].shape)
print(weights[1].shape)
```

## Output shape

```python
import tensorflow as tf
import tensorflow.keras as keras

# create a conv layer
a_conv_layer = keras.layers.Conv2D(filters=1, kernel_size=(2, 2))

# create input data
input_data = tf.random.uniform((1, 3, 3, 1), minval=0, maxval=1)

# call 'a_conv_layer' function
output_data = a_conv_layer(input_data)

print('input_data: \n', input_data.shape)
print('output_data: \n', output_data.shape)
```

```python
import tensorflow as tf
import tensorflow.keras as keras

# create a conv layer
a_conv_layer = keras.layers.Conv2D(filters=32, kernel_size=(2, 2))

# create input data
input_data = tf.random.uniform((1, 3, 3, 2), minval=0, maxval=1)

# call 'a_conv_layer' function
output_data = a_conv_layer(input_data)

print('input_data: \n', input_data.shape)
print('output_data: \n', output_data.shape)
```
